```python
import tensorflow.keras 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,shear_range=0.2,
                                 rotation_range=180,zoom_range=0.2,
                                horizontal_flip=True)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
x_train=train_datagen.flow_from_directory(r'E:\InternshipProject\Dataset\Dataset\train_set',
                                         target_size=(128,128),
                                         batch_size=32,
                                         class_mode='binary')
x_test=test_datagen.flow_from_directory(r'E:\InternshipProject\Dataset\Dataset\test_set',
                                         target_size=(128,128),
                                         batch_size=32,
                                         class_mode='binary')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
import warnings
warnings.filterwarnings('ignore')
```

```python
model=Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape=(128,128,3),activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units=150,activation='relu'))
```

```python
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy',
                optimizer="adam",
                metrics=["accuracy"])
```

```python
model.fit_generator(x_train,steps_per_epoch=14,
                   epochs=15,validation_data=x_test,
                   validation_steps=4)
```

```python
model.save("forest1.h5")
```
